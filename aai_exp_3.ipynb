```python
import numpy as np
from sklearn.mixture import GaussianMixture

# Sample data (replace with your own dataset)
X = np.random.rand(100, 2)

# Create and fit the GMM model
n_components = 2  # Number of Gaussian components (clusters)
gmm = GaussianMixture(n_components=n_components)  # Fixed syntax error here
gmm.fit(X)

# Predict the cluster assignments for data points
cluster_assignments = gmm.predict(X)

# Predict the likelihood of data points belonging to each cluster
cluster_probabilities = gmm.predict_proba(X)

print("Cluster Assignments:")
print(cluster_assignments)
print("Cluster Probabilities:")
print(cluster_probabilities)
```
